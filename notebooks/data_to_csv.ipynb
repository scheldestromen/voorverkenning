# Save timeseries as CSV

```python
import hydropandas as hpd
from tqdm import tqdm
import os
```

```python
hpd.__version__
```

# Load data

```python
DATA_DIR = r"../data//"
EXPORT_DIR = r"export//"
```

```python
oc = hpd.read_pickle(DATA_DIR + "oc_unsat.pkl")
# renamed to oc_unsat.pkl in c:\data\python\cloned\voorverkenning\data
```

# Dump to CSV

```python
# meta data
oc.drop(columns=['obs']).to_csv(os.path.join(EXPORT_DIR, "meta.csv"), index=True)
```

```python
# timeseries
for idx, row in tqdm(oc.iterrows(), total=len(oc), desc="Saving timeseries"):
    # Convert obs DataFrame to CSV, file name is index value
    try:
        row.obs.to_csv(os.path.join(EXPORT_DIR, f"{idx}.csv"), index=True)
    except Exception as e:
        print(f"Error saving {idx}: {e}")
```

```python
# move all csv files in export directory to zip file
import zipfile
with zipfile.ZipFile(os.path.join(EXPORT_DIR, "timeseries.zip"), 'w', compression=zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(EXPORT_DIR):
        for file in files:
            if file.endswith('.csv'):
                zipf.write(os.path.join(root, file), arcname=file)
                os.remove(os.path.join(root, file))
```
